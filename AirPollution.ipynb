```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dfx = pd.read_csv("C:/Users/ASUS/Desktop/p/ML/ML/Dataset/multivariat_regression_airPollution/Train.csv")
```

```python
X = dfx.values[:,:-1]
Y = dfx.values[:,-1]
Y = np.reshape(Y,(-1,1))
print(X)
print(Y.shape)
```

```python
ones = np.ones((X.shape[0],1))
X = np.hstack((ones,X))
```

```python
theta = np.dot(np.linalg.pinv(np.dot(X.T,X)),np.dot(X.T,Y))
print(theta)
```

```python
test = pd.read_csv("C:/Users/ASUS/Desktop/p/ML/ML/Dataset/multivariat_regression_airPollution/Test.csv")
```

```python
test = test.values
print(test.shape)
```

```python
def prediction(X_test,theta):
    out = theta[0] + theta[1]*X_test[0] + theta[2]*X_test[1] + theta[3]*X_test[2] + theta[4]*X_test[3] + theta[5]*X_test[4]
    return out
```

```python
print(test[0])
```

```python
m = test.shape[0]
output = []
print(test[0].shape)
for i in range(m):
    out = prediction(test[i],theta)
    print(out)
    output.append(out[0])
# output = np.reshape(output,(-1,1))
# print(output)
    
```

```python
y =  pd.DataFrame(data=output,columns=["target"]) 
y.to_csv("AirPullution0.csv",index = "Id")
```
