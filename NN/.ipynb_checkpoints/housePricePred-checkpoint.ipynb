```python
import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("./Dataset/housePricePred/train.csv")

test = pd.read_csv("./Dataset/housePricePred/Test_Data.csv")
```

```python
train
```

```python
columnsToRemove = ['Id','Alley','PoolQC','MiscFeature','Fence']

train.drop(columnsToRemove,axis = 1,inplace=True)

test.drop(columnsToRemove,axis = 1,inplace=True)
# test.drop('Alley',axis = 1, inplace=True)

train.fillna('0',inplace=True)
test.fillna('0',inplace=True)

train = train.apply(pd.to_numeric, errors='ignore')
```

```python
for i in range(1460):
    print(train['LotFrontage'][i])
```

```python
train.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le  = LabelEncoder()
labelsToEncode = ['']
```

```python
train
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
# remove all feilds with no integer

train = train.select_dtypes(exclude=['object'])
train.drop('Id',axis = 1,inplace=True)
train.fillna(0,inplace=True)

test = test.select_dtypes(exclude=['object'])
ID = test.Id
test.fillna(0,inplace=True)
test.drop('Id',axis = 1,inplace=True)
```

```python
from sklearn.ensemble import IsolationForest
```

```python
clf = IsolationForest(max_samples = 100, random_state = 42)
clf.fit(train)
y_noano = clf.predict(train)
y_noano = pd.DataFrame(y_noano, columns = ['Top'])
y_noano[y_noano['Top'] == 1].index.values

train = train.iloc[y_noano[y_noano['Top'] == 1].index.values]
train.reset_index(drop = True, inplace = True)
print("No of outliers ", y_noano[y_noano['Top'] == -1].shape[0])
print(train.shape[0])
```

```python
train.head(10)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
col_train = list(train.columns)
col_train_bis = list(train.columns)

col_train_bis.remove('SalePrice')

mat_train = np.matrix(train)
mat_test = np.matrix(test)
mat_new = np.matrix(train.drop('SalePrice',axis=1))
mat_y = np.array(train.SalePrice).reshape((1314,1))

prepro_y = MinMaxScaler()
prepro_y.fit(mat_y)

prepro = MinMaxScaler()
prepro.fit(mat_train)

prepro_test = MinMaxScaler()
prepro_test.fit(mat_test)

train = pd.DataFrame(prepro.transform(mat_train),columns=col_train)
test = pd.DataFrame(prepro_test.transform(mat_test),columns=col_train_bis)
```

```python
test.shape
```

```python
train = np.array(train)

X_train = train[:,:-1]
Y_train = train[:,-1]
Y_train = np.reshape(Y_train,(-1,1))
```

```python
# neural network
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(200,activation='relu',input_dim=36,kernel_initializer='normal'))
model.add(Dense(100,activation='relu',kernel_initializer='normal'))
model.add(Dense(50,activation='relu',kernel_initializer='normal'))
model.add(Dense(25,activation='relu',kernel_initializer='normal'))
model.add(Dense(1,kernel_initializer='normal'))

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.summary()
```

```python
hist = model.fit(X_train,Y_train,epochs=550,batch_size=8)
```

```python
model.evaluate(X_train,Y_train)
```

```python
output = model.predict(test)
```

```python
predictions = prepro_y.inverse_transform(np.array(output).reshape(360,1))
```

```python
print(predictions)
```

```python
import matplotlib
import matplotlib.pyplot as plt
from pylab import rcParams
matplotlib.rc('xtick', labelsize=30) 
matplotlib.rc('ytick', labelsize=30) 

fig, ax = plt.subplots(figsize=(50, 40))

plt.style.use('ggplot')
plt.plot(predictions, 'ro')
plt.xlabel('Predictions', fontsize = 30)
plt.ylabel('Reality', fontsize = 30)
plt.title('Predictions x Reality on dataset Test', fontsize = 30)
# ax.plot([reality.min(), reality.max()], [reality.min(), reality.max()], 'k--', lw=4)
plt.show()
```
